```python
import os
import cv2
import numpy as np
from keras.preprocessing.image import ImageDataGenerator
```

```python
input_folder = '/content/drive/MyDrive/Road damage detection/Thesis/Reveling'
output_folder = '/content/drive/MyDrive/Road damage detection/augmented/Reveling'
```

```python
datagen = ImageDataGenerator(
    rotation_range=50,        # Random rotation between 0-20 degrees
    width_shift_range=0.2,    # Random horizontal shift of up to 20% of the width
    height_shift_range=0.2,   # Random vertical shift of up to 20% of the height
    zoom_range=0.3,           # Random zoom of up to 30%
    horizontal_flip=True,     # Randomly flip images horizontally
    vertical_flip=False,      # Do not flip images vertically
    fill_mode='nearest'
)
```

```python
image_files = os.listdir(input_folder)
for file_name in image_files:
  image_path = os.path.join(input_folder, file_name)
  image = cv2.imread(image_path)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Resize the image
  target_size = (480, 480)  # Set the desired target size
  image = cv2.resize(image, target_size)
      # Expand the dimensions of the image to match the expected input shape of the generator
  image = np.expand_dims(image, axis=0)

    # Generate augmented images
  augmented_images = datagen.flow(image, batch_size=1)

    # Save the augmented images
  for i, augmented_image in enumerate(augmented_images):
        augmented_image = augmented_image[0].astype(np.uint8)
        augmented_image = cv2.cvtColor(augmented_image, cv2.COLOR_RGB2BGR)
        output_path = os.path.join(output_folder, f"'all_{file_name.split('.')[0]}_{i}.jpg")
        cv2.imwrite(output_path, augmented_image)

        if i == 9:  # Adjust this number as desired
            break

```
